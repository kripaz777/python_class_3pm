```python
# # Control statemant
# break
# continue
```

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
for i in range(10):
    if i == 5:
        continue
    print(i, end = " ")
```

```python
a = "Hello World"
for i in a:
    if i == " ":
        break
    print(i,end = "")
```

```python
a = "Hello World I am Python. I am best"
for i in a:
    if i == ".":
        break
    print(i,end = "")
print(".")
```

```python
a = "Hello World I am Python. I am best"
for i in a:
    if i == ".":
        continue
    print(i,end = "")
```

```python
#break in while
i = 0
while i <= 10:
    if i == 5:
        break
    print(i)
    i = i+1
```

```python
#continue in while
i = 0
while i <= 10:
    if i == 5:
        i = i+1
        continue
    print(i, end = " ")
    i = i+1
```
